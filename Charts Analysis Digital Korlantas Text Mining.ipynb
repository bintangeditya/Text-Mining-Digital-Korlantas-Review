```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_data():
    data = pd.read_csv("digital_korlantas_reviews_5000_preprocessed.csv")
    return data

df = load_data()
df = df[["content_preprocessed","sentiment"]] #hanya mengambil kolom content_preprocessed dan sentiment 
df.head()
```

```python
dfPositif = df[df["sentiment"]=="Positif"]
dfNegatif = df[df["sentiment"]=="Negatif"]
print("All :",len(df))
print("Positif :",len(dfPositif))
print("Negatif :",len(dfNegatif))
```

```python
word_count_all = df.content_preprocessed.str.split(expand=True).stack().value_counts()
word_count_all = word_count_all.to_dict()
# dict(list(word_count_all.items())[0:10])
word_count_positif = dfPositif.content_preprocessed.str.split(expand=True).stack().value_counts()
word_count_positif = word_count_positif.to_dict()
# dict(list(word_count_positif.items())[0:10])
word_count_negatif = dfNegatif.content_preprocessed.str.split(expand=True).stack().value_counts()
word_count_negatif = word_count_negatif.to_dict()
# dict(list(word_count_negatif.items())[0:10])
```
